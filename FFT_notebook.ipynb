```python
from astropy.modeling import models
import numpy as np
from pylab import *
import matplotlib
```

```python
# timestamps
x=np.arange(0,10*np.pi,.003)
# eqv frequency
freq = np.fft.rfftfreq(len(x),d=0.003)
```

```python
# signal is sum of three sines
f=models.Sine1D(amplitude=1.,frequency=1.)+models.Sine1D(amplitude=.6,frequency=3.)+models.Sine1D(amplitude=.6,frequency=2.)
```

```python
plot(x,f(x))
xlim(0,5.)
xlabel("time ($t$)")
ylabel("signal $f(t)$")
```

```python
fft=np.fft.rfft(f(x))
plot(freq,abs(fft*fft))
xlabel("frequency ($\omega$)")
ylabel(r"FFT power $| \tilde{f} (\omega)|^2$")
axis([0,5,0,1.7e7])
```

```python
# some noise
y=f(x)+np.random.normal(0.,.3,len(x))
fft=np.fft.rfft(y)

plot(x,y)
xlabel("time ($t$)")
ylabel("signal $f(t)$")
xlim(0.,5.)
```

```python

plot(freq,abs(fft*fft))
xlabel("frequency ($\omega$)")
ylabel(r"FFT power $| \tilde{f} (\omega)|^2$")
axis([0,5,0,1.7e7])
```

```python
# some more noise
y=f(x)+np.random.normal(0.,6.,len(x))
fft=np.fft.rfft(y)

plot(x,y)
xlabel("time ($t$)")
ylabel("signal $f(t)$")
xlim([0,5])
```

```python

plot(freq,abs(fft*fft))
xlabel("frequency ($\omega$)")
ylabel(r"FFT power $| \tilde{f} (\omega)|^2$")
axis([0,5,0,3.e7])
```

```python
f=models.Sine1D(amplitude=1.,frequency=1.)+models.Sine1D(amplitude=.6,frequency=3.)+models.Sine1D(amplitude=.6,frequency=2.)+models.Sine1D(amplitude=3.0,frequency=4.1)
```

```python
# add some really bad interference
y=f(x)+np.random.normal(0.,.5,len(x))
fft=np.fft.rfft(y)
plot(x,y)
xlabel("time ($t$)")
ylabel("signal $f(t)$")
xlim([0,5])
```

```python
plot(freq,abs(fft*fft))
xlabel("frequency ($\omega$)")
ylabel(r"FFT power $| \tilde{f} (\omega)|^2$")
axis([0,5,0,1.7e8])
```

```python
fft[np.where((freq>3.5) & (freq<4.5))]=0.
```

```python
plot(freq,abs(fft*fft))
xlabel("frequency ($\omega$)")
ylabel(r"FFT power $| \tilde{f} (\omega)|^2$")
axis([0,5,0,1.7e7])
```
